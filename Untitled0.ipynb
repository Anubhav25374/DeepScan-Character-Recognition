```python
# Step 2: Import all necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
from PIL import Image
import string
import os
```

```python
# Step 3: Download EMNIST Letters dataset (A-Z characters)
!pip install tensorflow-datasets
import tensorflow_datasets as tfds

# Load EMNIST letters dataset
(ds_train, ds_test), ds_info = tfds.load(
    'emnist/letters',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True
)

print("Dataset Info:")
print(ds_info)
print(f"\nTotal training examples: {ds_info.splits['train'].num_examples}")
print(f"Total test examples: {ds_info.splits['test'].num_examples}")
```

```python
# Step 4: Preprocess images - normalize and reshape

IMG_SIZE = 28
BATCH_SIZE = 128
NUM_CLASSES = 26  # A-Z

def preprocess(image, label):
    image = tf.cast(image, tf.float32) / 255.0          # Normalize to [0, 1]
    image = tf.transpose(image, perm=[1, 0, 2])         # Fix EMNIST rotation
    label = label - 1                                    # Labels are 1-26, shift to 0-25
    return image, label

# Apply preprocessing
train_ds = ds_train.map(preprocess).shuffle(10000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_ds  = ds_test.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Label map: 0 = 'A', 1 = 'B', ..., 25 = 'Z'
label_map = {i: chr(65 + i) for i in range(26)}
print("Label Map:", label_map)
```

```python
# Step 5: Visualize some training examples

fig, axes = plt.subplots(2, 8, figsize=(16, 4))
for images, labels in train_ds.take(1):
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i].numpy().squeeze(), cmap='gray')
        ax.set_title(f"Label: {label_map[labels[i].numpy()]}", fontsize=10)
        ax.axis('off')
plt.suptitle("Sample Training Images", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Build a Convolutional Neural Network (CNN)

model = keras.Sequential([
    keras.Input(shape=(28, 28, 1)),

    # Block 1
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Block 2
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Classifier Head
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

model.summary()
```

```python
# Step 7: Compile with optimizer and loss

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Step 8: Train the model with callbacks

callbacks = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3, min_lr=1e-6)
]

history = model.fit(
    train_ds,
    epochs=20,
    validation_data=test_ds,
    callbacks=callbacks
)
```

```python
# Step 9: Visualize training accuracy and loss

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history.history['accuracy'], label='Train Accuracy')
ax1.plot(history.history['val_accuracy'], label='Val Accuracy')
ax1.set_title('Model Accuracy')
ax1.set_xlabel('Epoch')
ax1.legend()

ax2.plot(history.history['loss'], label='Train Loss')
ax2.plot(history.history['val_loss'], label='Val Loss')
ax2.set_title('Model Loss')
ax2.set_xlabel('Epoch')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 10: Final evaluation on test data

test_loss, test_acc = model.evaluate(test_ds)
print(f"\n✅ Test Accuracy: {test_acc * 100:.2f}%")
print(f"📉 Test Loss: {test_loss:.4f}")
```

```python
# Step 11: Predict character from your own uploaded image

from google.colab import files
from IPython.display import display

def predict_character(image_path):
    """Load an image, preprocess it, and predict the English character."""
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img = img.resize((28, 28))                  # Resize to 28x28
    img_array = np.array(img)

    # Auto-invert if background is white (like handwritten on white paper)
    if img_array.mean() > 127:
        img_array = 255 - img_array

    img_array = img_array.astype('float32') / 255.0
    img_array = img_array.reshape(1, 28, 28, 1)

    # Predict
    predictions = model.predict(img_array)
    predicted_index = np.argmax(predictions)
    confidence = predictions[0][predicted_index] * 100

    # Show image
    plt.figure(figsize=(3, 3))
    plt.imshow(img, cmap='gray')
    plt.title(f"Predicted: {label_map[predicted_index]}\nConfidence: {confidence:.1f}%", fontsize=13)
    plt.axis('off')
    plt.show()

    print(f"\n🔤 Predicted Character : {label_map[predicted_index]}")
    print(f"📊 Confidence          : {confidence:.2f}%")

# Upload your own image
print("Upload an image of a handwritten English character:")
uploaded = files.upload()

for filename in uploaded.keys():
    predict_character(filename)
```
